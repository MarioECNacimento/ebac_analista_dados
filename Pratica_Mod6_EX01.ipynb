```python
#Imporando as bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
#montando o dataframe
sinascRO = pd.read_csv('SINASC_RO_2019.csv', sep=',')
```

```python
#visualizando o dataframe 
sinascRO.head()
```

```python
#Visualizando as informações das colunas do dataframe
sinascRO.info()
```

```python
#Copiando e selecionado as colunas que serão trabalhadas nos exercicios 
sinascRO_clone = sinascRO[:]
sinascRO_final = sinascRO_clone[['APGAR1', 'APGAR5', 'SEXO', 'GESTACAO', 'ESTCIVMAE', 'CONSULTAS']]
sinascRO_final.head()
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
plt.rc('figure', figsize=(12, 5))
fig, axes = plt.subplots(1, 2)
fig1 = sns.countplot(ax = axes[0], x='APGAR1', data=sinascRO_final, palette='Set1')
fig2 = sns.countplot(ax = axes[1], x='APGAR5', data=sinascRO_final, palette='Set1')
plt.subplots_adjust(wspace=0.5, hspace=0)

fig1.set_xticklabels(fig1.get_xticklabels(), rotation=45, horizontalalignment='right')
fig1.set_title('Apgar no Primeiro Minuto')
fig2.set_xticklabels(fig2.get_xticklabels(), rotation=45, horizontalalignment='right')
fig2.set_title('Apgar no Quinto Minuto')
None
```

```python
plt.rc('figure', figsize=(14, 5))
fig, axes = plt.subplots(1, 3)
fig1 = sns.countplot(ax = axes[0], x='SEXO', data=sinascRO_final, palette='Set1')
fig2 = sns.countplot(ax = axes[1], x='GESTACAO', data=sinascRO_final, palette='Set1')
fig3 = sns.countplot(ax = axes[2], x='ESTCIVMAE', data=sinascRO_final, palette='Set1')
plt.subplots_adjust(wspace=0.5, hspace=0)

fig1.set_xticklabels(fig1.get_xticklabels(), rotation=45, horizontalalignment='right')
fig1.set_title('Sexo')
fig2.set_xticklabels(fig2.get_xticklabels(), rotation=45, horizontalalignment='right')
fig2.set_title('Gestacão em Semanas')
fig3.set_xticklabels(fig3.get_xticklabels(), rotation=45, horizontalalignment='right')
fig3.set_title('Estado Civil da Mãe')
None
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
apgar5_mean = [round(sinascRO_final['APGAR5'][sinascRO_final['CONSULTAS']==1].mean(), 2),
       round(sinascRO_final['APGAR5'][sinascRO_final['CONSULTAS']==2].mean(), 2),
       round(sinascRO_final['APGAR5'][sinascRO_final['CONSULTAS']==3].mean(), 2),
       round(sinascRO_final['APGAR5'][sinascRO_final['CONSULTAS']==4].mean(), 2),
       round(sinascRO_final['APGAR5'][sinascRO_final['CONSULTAS']==9].mean(), 2)]
consultas = ['Nenhuma', '1 - 3', '4 - 6', '7 ou Mais', 'Ignorado']
plt.title('APGAR 5 Médio por Consultas')
plt.xlabel('Consultas')
plt.ylabel('APGAR 5 Médio')
plt.bar(consultas, apgar5_mean, color='blue')
plt.show()
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
plt.rc('figure', figsize=(12, 12))
fig, axes = plt.subplots(2, 2)
fig1 = sns.countplot(ax = axes[0, 0], x='SEXO', data=sinascRO_final, palette='Set1')
fig2 = sns.countplot(ax = axes[0, 1], x='GESTACAO', data=sinascRO_final, palette='Set1')
fig3 = sns.countplot(ax = axes[1, 0], x='ESTCIVMAE', data=sinascRO_final, palette='Set1')
fig4 = sns.countplot(ax = axes[1, 1], x='CONSULTAS', data=sinascRO_final, palette='Set1')
plt.subplots_adjust(wspace=0.5, hspace=0.7)

fig1.set_xticklabels(fig1.get_xticklabels(), rotation=45, horizontalalignment='right')
fig1.set_title('Sexo')
fig2.set_xticklabels(fig2.get_xticklabels(), rotation=45, horizontalalignment='right')
fig2.set_title('Gestacão em Semanas')
fig3.set_xticklabels(fig3.get_xticklabels(), rotation=45, horizontalalignment='right')
fig3.set_title('Estado Civil da Mãe')
fig4.set_xticklabels(fig4.get_xticklabels(), rotation=45, horizontalalignment='right')
fig4.set_title('Consultas')
None
```
